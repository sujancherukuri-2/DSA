<h2> 1. MAXIMUM CONSECUTIVE ONES </h2>

<h3> <u> BruteForce </u></h3>

<p>Start at each index i.<br>

If a[i] == 1, start counting consecutive 1s.

Move forward until you find 0.

Update the maximum.

Repeat for all indices.</p>

<h4> TC - O(N^2) , SC - O(N) </h4>

```python
// Brute Force
        for (int i = 0; i < n; i++) 
        {
            int count = 0;

            for (int j = i; j < n; j++) 
            {
                if (a[j] == 1) 
                {
                    count++;
                    maxi = Math.max(maxi, count);
                } 
                else 
                {
                    break;
                }
            }
        }
```

<h3> <u> Optimal Approach </u></h3>

<li> Read array.<br>

If element is 1 → increase count.

If element is 0 → reset count to 0.

Keep updating maximum count.</li>

<h4> TC - O(N) , SC - O(1)</h4>

```python
import java.util.*;

        Scanner sc = new Scanner(System.in);

        int n = sc.nextInt();
        int[] a = new int[n];
        int c = 0;
        int maxi = Integer.MIN_VALUE;

        for (int i = 0; i < n; i++) {
            a[i] = sc.nextInt();
        }

        for (int i = 0; i < n; i++) {
            if (a[i] == 1) {
                c++;
                maxi = Math.max(maxi, c);
            } else {
                c = 0;
            }
        }

        System.out.println(maxi);
```
