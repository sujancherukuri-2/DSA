<h2> 2. MOVE ZEROES TO END </h2>

Input: 1 , 0 , 2 , 3 , 0 , 4 , 0 , 1 <br>
Output: 1 , 2 , 3 , 4 , 1 , 0 , 0 , 0

<h3> <u>Bruteforce</u></h3>

<p>Create a temporary array and set k = 0.<br>
Traverse the array and store all non-zero elements in temp[k], increment k.<br>
Copy the first k elements from temp back to arr.<br>
Sort the first k elements of arr.<br>
Fill the remaining positions of arr with 0.</p>

```python
import java.util.*;
Scanner sc = new Scanner(System.in);

int n = sc.nextInt();
int[] arr = new int[n];
for(int i=0; i<n; i++){
    arr[i] = sc.nextInt();
}
int k=0;
int temp[] = new int[n];

for(int i=0; i<n; i++)
{
    if(arr[i]!=0)
    {
        temp[k++]=arr[i];
    }
}
for(int i=0;i<k;i++)
{
    arr[i]=temp[i];
}
Arrays.sort(arr,0,k);
for(int i=k;i<n;i++)
{
    arr[i]=0;
}
for(int i=0;i<n;i++)
{
    System.out.print(arr[i]+" ");
}
```

<h3> <u>Optimal Approach</u></h3>

<h4> Use a two-pointer approach where j tracks the position of the next non-zero element and swap whenever a non-zero element is found during traversal.</h4>

<li>Initialize j = 0.<br>
Traverse the array from i = 0 to n-1.<br>
If nums[i] is not 0:<br>
Swap nums[i] with nums[j].<br>
Increment j.<br>
Continue until the array ends.</li>

```python
import java.util.*;
Scanner sc = new Scanner(System.in);

        int n = sc.nextInt();
        int[] nums = new int[n];

        for(int i = 0; i < n; i++){
            nums[i] = sc.nextInt();
        }

        int j = 0;

        // Swap of non zero elements with zero elements
        for(int i = 0; i < nums.length; i++)
        {
            if(nums[i] != 0)
            {
                int temp = nums[i];
                nums[i] = nums[j];
                nums[j] = temp;
                j++;
            }
        }

        for(int i = 0; i < n; i++)
        {
            System.out.print(nums[i] + " ");
        }
       
```
