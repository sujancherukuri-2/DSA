<h2> LEFT ROTATE ARRAY BY K PLACES </h2>

<p> Input : - arr = [1,2,3,4,5,6,7] , k = 2 , n = 7

Output : - [3,4,5,6,7,1,2]</p>

<h3> <b><u>1.BruteForce</b></u></h3>

<p> <b>Idea: Rotate the array left by 1 exactly k times.</b></p>

<p>Repeat k times:<br>

Store the first element in temp.

Shift all elements one position to the left.

Put temp at the last position</p>

<h4> TC - O(N * K) , SC - O(1) </h4>

```python
import java.util.*;
Scanner sc = new Scanner(System.in);

int n = sc.nextInt();
int[] arr = new int[n];
for(int i = 0; i < n; i++)
    {
        arr[i] = sc.nextInt();
    }
int k = sc.nextInt();
for(int i = 0; i < k; i++)
    {
        int temp = arr[0];
        for(int j = 1; j < n; j++)
        {
            arr[j-1] = arr[j];
        }
        arr[n-1] = temp;
    }
for(int i = 0; i < n; i++)
    {
        System.out.print(arr[i] + " ");
    }
```

<h3> <u><b>2.Better Approach</u></b></h3>

<p> <b>Idea: Store first k elements, shift remaining, then add stored elements at end.</b></p>

<h4> TC - O(N) , SC - O(K) </h4>

```python
k = k % n;

int[] temp = new int[k];

for(int i = 0; i < k; i++){
    temp[i] = arr[i];
}

for(int i = k; i < n; i++){
    arr[i-k] = arr[i];
}

for(int i = 0; i < k; i++){
    arr[n-k+i] = temp[i];
}

System.out.println("Rotated Array:");
for(int i = 0; i < n; i++){
    System.out.print(arr[i] + " ");
}
```

<h3><u><b>3.Optimal Approach </u></b></h3>

<p>Reverse first k elements<br>

Reverse n-k elements

Reverse entire array</p>

<h4> TC - O(N) , SC - O(1) </h4>

```python
import java.util.*;

public class Main {

    public static void reverse(int[] arr, int start, int end){
        while(start < end){
            int temp = arr[start];
            arr[start] = arr[end];
            arr[end] = temp;
            start++;
            end--;
        }
    }

    public static void main(String[] args) {

        Scanner sc = new Scanner(System.in);
        int n = sc.nextInt();
        int[] arr = new int[n];
        for(int i = 0; i < n; i++){
            arr[i] = sc.nextInt();
        }

        int k = sc.nextInt();

        k = k % n;

        reverse(arr, 0, k-1);
        reverse(arr, k, n-1);
        reverse(arr, 0, n-1);

        for(int i = 0; i < n; i++){
            System.out.print(arr[i] + " ");
        }
    }
}
```
