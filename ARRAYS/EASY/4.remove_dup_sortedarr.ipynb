<h2> 4. REMOVE DUPLICATES IN PLACE SORTED ARRAY </h2>

<h3> <u><b>1. Bruteforce</u></b></h3>

<p>Create a HashSet to track seen elements.<br>

Traverse the array.

If element not in set:

Add it to set

Store it in the array

Move index

Return number of unique elements.</p>

<h4> TC - O(N) , SC - O(N) </h4>

```python
        Scanner sc = new Scanner(System.in);
        int n = sc.nextInt();
        int nums[] = new int[n];
        for (int i = 0; i < n; i++) {
            nums[i] = sc.nextInt();
        }

        HashSet<Integer> seen = new HashSet<>();
        int index = 0;

        for (int i = 0; i < n; i++) {
            int num = nums[i];

            if (!seen.contains(num)) {
                seen.add(num);
                nums[index] = num;
                index++;
            }
        }

        for (int i = 0; i < index; i++) {
            System.out.print(nums[i] + " ");
        }
        System.out.println("\nUnique count: " + index);
```

<h3> <u><b>2. Optimal Approach </u></b></h3>

<p>Assume the first element is unique → set i = 0.<br>

Start checking elements from the second element (j = 1).

If the current element is different from the last unique element (nums[j] != nums[i]):

Move i one step forward.

Copy the current element to position i.

Continue this process until the array ends.

The number of unique elements = i + 1.</p>

<h4> TC - O(N) , SC - O(1) </h4>

```python
        Scanner sc = new Scanner(System.in);
        int n = sc.nextInt();
        int nums[] = new int[n];
        for (int i = 0; i < n; i++) {
            nums[i] = sc.nextInt();
        }
        int i = 0;

        for (int j = 1; j < nums.length; j++) {
            if (nums[j] != nums[i]) {
                i++;
                nums[i] = nums[j];
            }
        }

        int k = i + 1;

        for (int x = 0; x < k; x++) {
            System.out.print(nums[x] + " ");
        }
        System.out.println("\nUnique count: " + k);
```
