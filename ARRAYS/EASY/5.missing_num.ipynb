<h2> 5. MISSING NUMBER </h2>

<h3> <u><b> Bruteforce Approach</u></b></h3>

<p>Loop from 0 → n</br>

For each number, search it in the array

If not found → that is the missing number</p>

<h4> TC - O(N^2) , SC - O(1) </h4>

```python
        Scanner sc = new Scanner(System.in);
        int n = sc.nextInt();
        int arr[] = new int[n];

        for(int i = 0; i < n; i++){
            arr[i] = sc.nextInt();
        }

        for(int i = 0; i <= n; i++){

            boolean found = false;

            for(int j = 0; j < n; j++){
                if(arr[j] == i){
                    found = true;
                    break;
                }
            }

            if(!found){
                System.out.println("Missing number: " + i);
                break;
            }
        }
```

<h3><u><b> HASHING APPROACH </u></b></h3>

<p>Read n and array elements.<br>

 Create a HashMap.

 Store every array element in the map.

 Check numbers from 0 to n:

If a number is not in the map → it is the missing number.

 Print the missing number.<p>

 <h3> TC - O(N) , SC - O(N) </h3>

```python
        Scanner sc = new Scanner(System.in);
        int n = sc.nextInt();
        int arr[] = new int[n];
        HashMap<Integer, Integer> map = new HashMap<>();

        // store elements in map
        for(int i = 0; i < n; i++){
            arr[i] = sc.nextInt();
            map.put(arr[i], map.getOrDefault(arr[i],0) + 1);
        }

        // find missing number
        for(int i = 0; i <= n; i++){
            if(map.get(i) == null){
                System.out.println("Missing number: " + i);
                break;
            }
        }
```

<h4> SUM / XOR APPROACH</h4>

SUM APP -> N*(N+1)/2 - (SUM OF ALL ELEM)

<p>Initialize two XOR variables:<br>

xor1 → XOR of numbers 0 → n

xor2 → XOR of array elements

XOR both results.</p>

<h3> TC - O(N) , SC - O(1) </h3>

```python
        Scanner sc = new Scanner(System.in);
        int n = sc.nextInt();
        int arr[] = new int[n];

        int xor1 = 0;
        int xor2 = 0;

        // XOR of array elements
        for(int i = 0; i < n; i++){
            arr[i] = sc.nextInt();
            xor2 = xor2 ^ arr[i];
        }

        // XOR of numbers from 0 to n
        for(int i = 0; i <= n; i++){
            xor1 = xor1 ^ i;
        }

        int missing = xor1 ^ xor2;

        System.out.println("Missing number: " + missing);
```
