<h2> 6. FIND NO. APPEARING ONCE , WHERE ALL ARE TWICE </h2>

<h3><u><b>1.Bruteforce</u></b></h3>

<p>Traverse the array.<br>

For each element, count its occurrences using another loop.

If count = 1 → print that element.</p>

<h4> TC - O(N^2) , SC = O(1) </h4>

```python
        Scanner sc = new Scanner(System.in);
        int n = sc.nextInt();
        int arr[] = new int[n];

        for(int i = 0; i < n; i++){
            arr[i] = sc.nextInt();
        }

        for(int i = 0; i < n; i++){
            int count = 0;
            for(int j = 0; j < n; j++){
                if(arr[i] == arr[j]){
                    count++;
                }
            }

            if(count == 1){
                System.out.println("Single element: " + arr[i]);
                break;
            }
        }
```

<h3><u><b>2. HASHING </u></b></h3>

<p>Store element frequencies in HashMap.<br>

Traverse the map.

Element with frequency = 1 is the answer.</p>

<h4> TC - O(N) , SC - O(N) </h4>

```python
        Scanner sc = new Scanner(System.in);
        int n = sc.nextInt();
        int arr[] = new int[n];

        HashMap<Integer, Integer> map = new HashMap<>();

        // input array
        for(int i = 0; i < n; i++){
            arr[i] = sc.nextInt();
        }

        // store frequency using for-each
        for(int x : arr){
            map.put(x, map.getOrDefault(x, 0) + 1);
        }

        // find single element
        for(int x : arr){
            if(map.get(x) == 1){
                System.out.println("Single element: " + x);
                break;
            }
        }
```

<h3><u><b>3.Optimal</u></b></h3>

<p>Initialize xor = 0<br>

Traverse array

XOR every element

Result = single element</p>

<h4> TC - O(N) , SC-O(1) </h4>

```python
        Scanner sc = new Scanner(System.in);
        int n = sc.nextInt();
        int arr[] = new int[n];

        int xor = 0;

        for(int i = 0; i < n; i++){
            arr[i] = sc.nextInt();
            xor = xor ^ arr[i];
        }

        System.out.println("Single element: " + xor);
```
