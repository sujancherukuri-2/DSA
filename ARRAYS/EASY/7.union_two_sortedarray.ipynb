<h2> 7. UNION OF TWO SORTED ARRAYS </h2>

Input:n = 5,m = 5 arr1[] = {1,2,3,4,5}  arr2[] = {2,3,4,4,5}<br>
Output: {1,2,3,4,5}



<h4><u><b>1.HASHMAP</u></b></h4>

<p>Take n and m as input.<br>

Read arr1 and arr2.

Store elements in HashMap with frequency.

Traverse map keys → union elements.</p>

<h4> HASHMAP - TC - O(N + M) , SC = O(N + M) ---- NON SORTED<br>
TREEMAP - TC - O((N + M) LOG(N + M)) , SC = O(N + M) ---- SORTED</h4>

```python
        Scanner sc = new Scanner(System.in);

        // Input sizes
        int n = sc.nextInt();
        int m = sc.nextInt();

        int arr1[] = new int[n];
        int arr2[] = new int[m];

        // Input first array
        for(int i = 0; i < n; i++){
            arr1[i] = sc.nextInt();
        }

        // Input second array
        for(int i = 0; i < m; i++){
            arr2[i] = sc.nextInt();
        }

        HashMap<Integer, Integer> map = new HashMap<>();

        // Insert elements of arr1
        for(int i = 0; i < n; i++){
            map.put(arr1[i], map.getOrDefault(arr1[i], 0) + 1);
        }

        // Insert elements of arr2
        for(int i = 0; i < m; i++){
            map.put(arr2[i], map.getOrDefault(arr2[i], 0) + 1);
        }

        // Print union elements
        System.out.println("Union of arrays:");

        for(int key : map.keySet()){
            System.out.print(key + " ");
        }

        // HashSet<Integer> set = new HashSet<>();

        // // for-each loop for arr1
        // for(int x : arr1){
        //     set.add(x);
        // }

        // // for-each loop for arr2
        // for(int x : arr2){
        //     set.add(x);
        // }

        // System.out.println("Union:");

        // for(int x : set){
        //     System.out.print(x + " ");
        // }
```

<h4><u><b> 2.OPTIMAL - 2 POINTER </u></b></h4>

<p>Initialize two pointers i = 0, j = 0.<br>

Create a list union to store result.

While i < n and j < m

If arr1[i] <= arr2[j]

If element not already in union → add it

Increment i

Else

If element not already in union → add it

Increment j

Traverse remaining elements of arr1.

Traverse remaining elements of arr2.

Print union list.</p>

<h4> TC - O(N + M) , SC - O(N + M)</h4>

```python
        int i = 0, j = 0;

        ArrayList<Integer> union = new ArrayList<>();

        while(i < n && j < m){

            if(arr1[i] <= arr2[j]){

                if(union.size() == 0 || union.get(union.size()-1) != arr1[i]){
                    union.add(arr1[i]);
                }
                i++;

            } else {

                if(union.size() == 0 || union.get(union.size()-1) != arr2[j]){
                    union.add(arr2[j]);
                }
                j++;
            }
        }

        while(i < n){

            if(union.size() == 0 || union.get(union.size()-1) != arr1[i]){
                union.add(arr1[i]);
            }
            i++;
        }

        while(j < m){

            if(union.size() == 0 || union.get(union.size()-1) != arr2[j]){
                union.add(arr2[j]);
            }
            j++;
        }

        System.out.println(union);
```
