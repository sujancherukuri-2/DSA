<h2>8. TWO SUM </h2>

Input: N = 5, arr[] = {2,6,5,8,11}, target = 14<br>
Output : YES

<h3><u><b>1. Bruteforce</u></b></h3>

<p>Use two loops.<br>

First loop picks element i.

Second loop checks all elements j > i.

If nums[i] + nums[j] == target → return indices</p>

<h4> TC - O(N^2) , SC - O(1)</h4>

```python
        Scanner sc = new Scanner(System.in);
        int n = sc.nextInt();
        int nums[] = new int[n];
        for(int i = 0; i < n; i++){
            nums[i] = sc.nextInt();
        }
        int target = sc.nextInt();

        for(int i = 0; i < n; i++){
            for(int j = i + 1; j < n; j++){

                if(nums[i] + nums[j] == target){
                    System.out.println(i + " " + j);
                    //return;
                }
            }
        }
```

<h3><u><b>2. HASHMAP </u></b></h3>

<p>Create HashMap<br>

Traverse array

Find complement

complement = target - nums[i]

If complement exists in map → print indices

Otherwise store current element</p>

<h4> TC - O(N) , SC - O(N)</h4>

```python
        Scanner sc = new Scanner(System.in);
        int n = sc.nextInt();
        int nums[] = new int[n];
        for(int i = 0; i < n; i++){
            nums[i] = sc.nextInt();
        }

        int target = sc.nextInt();
        HashMap<Integer,Integer> map = new HashMap<>();

        for(int i = 0; i < n; i++){
            int comp = target - nums[i];
            if(map.containsKey(comp)){
                //System.out.println(map.get(comp) + " " + i);
                //return "YES";
                // return new int[] { map.get(comp), i }; - indices
            }
            map.put(nums[i], i);
        }
        //return NO
        //return new int[] { -1, -1 }; - indices
```

<h3><u><b>3. Optimal 2 pointer</u></b></h3>

<p>Sort array<br>

Use two pointers

left = 0
right = n-1

If sum < target → move left

If sum > target → move right

If equal → answer found</p>

<h4> TC - O(N LOGN) , SC - O(1) </h4>

```python
        Arrays.sort(nums);

        int left = 0;
        int right = n - 1;

        while(left < right){

            int sum = nums[left] + nums[right];

            if(sum == target){
                System.out.println(nums[left] + " " + nums[right]);
                return;
            }
            else if(sum < target){
                left++;
            }
            else{
                right--;
            }
        } 
```
