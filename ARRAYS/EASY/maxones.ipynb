<h2> 1. MAXIMUM CONSECUTIVE ONES </h2>

```python
import java.util.*;

        Scanner sc = new Scanner(System.in);

        int n = sc.nextInt();
        int[] a = new int[n];
        int c = 0;
        int maxi = Integer.MIN_VALUE;

        for (int i = 0; i < n; i++) {
            a[i] = sc.nextInt();
        }

        for (int i = 0; i < n; i++) {
            if (a[i] == 1) {
                c++;
                maxi = Math.max(maxi, c);
            } else {
                c = 0;
            }
        }

        System.out.println(maxi);
```
