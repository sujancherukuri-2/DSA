<h2> 2. MOVE ZEROES TO END </h2>

```python
import java.util.*;
Scanner sc = new Scanner(System.in);

int n = sc.nextInt();
int[] arr = new int[n];
for(int i=0; i<n; i++){
    arr[i] = sc.nextInt();
}
int k=0;
int temp[] = new int[n];

for(int i=0; i<n; i++)
{
    if(arr[i]!=0)
    {
        temp[k++]=arr[i];
    }
}
for(int i=0;i<k;i++)
{
    arr[i]=temp[i];
}
Arrays.sort(arr,0,k);
for(int i=k;i<n;i++)
{
    arr[i]=0;
}
for(int i=0;i<n;i++)
{
    System.out.print(arr[i]+" ");
}
```
