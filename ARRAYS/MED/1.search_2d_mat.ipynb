<h2>1. SEARCH IN 2D MATRIX</h2>

<h3><u><b> 1.Bruteforce </u></b></h3>

<p> Traverse matrix using two loops.<br>

Compare each element with target.

If equal → return true.

If not found → return false.</p>

<h4> TC - O(N^2) , SC - O(1)</h4>

```python
        int n = matrix.length;
        int m = matrix[0].length;

        boolean found = false;
        for(int i = 0; i < n; i++){
            for(int j = 0; j < m; j++){

                if(matrix[i][j] == target){
                    found = true;
                }
            }
        }
        System.out.println(found);
```

<h3><u><b> 2.Optimal </u></b></h3>

<p> 

```python
        int n = matrix.length;
        int m = matrix[0].length;

        int low = 0;
        int high = n*m - 1;

        boolean found = false;

        while(low <= high){

            int mid = (low + high) / 2;

            int row = mid / m;
            int col = mid % m;

            if(matrix[row][col] == target){
                found = true;
                break;
            }
            else if(matrix[row][col] < target){
                low = mid + 1;
            }
            else{
                high = mid - 1;
            }

        }

        System.out.println(found);
```
