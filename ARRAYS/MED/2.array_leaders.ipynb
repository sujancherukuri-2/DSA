<h2> 2. ARRAY LEADERS</h2>

<h3><u><b>1.Bruteforce</u></b></h3>

<p> Start from the first element.<br>

Assume the current element is a leader.

Compare it with every element to its right.

If any element ≥ current element, mark it not leader.

If it remains leader, store it in the answer list.

<h4> TC - O(N^2) , SC - O(1)</h4>

```python
        Scanner sc = new Scanner(System.in);
        int n = sc.nextInt();
        int arr[] = new int[n];

        for (int i = 0; i < n; i++) {
            arr[i] = sc.nextInt();
        }

        ArrayList<Integer> ans = new ArrayList<>();
        for (int i = 0; i < n; i++) {

            boolean leader = true;
            for (int j = i + 1; j < n; j++) {
                if (arr[j] >= arr[i]) {
                    leader = false;
                    break;
                }
            }
            if (leader) {
                ans.add(arr[i]);
            }
        }
        System.out.println(ans);
```

<h3><u><b>2. Optimal approach</u></b></h3>

<p>Rightmost element is always a leader.

Store the rightmost element in the answer list.

Set max = rightmost element.

Traverse the array from right to left.

If current element > max

add it to the list

update max = current element</p>

<h4> TC - O(N) , SC - O(1)</h4>

```python
        Scanner sc = new Scanner(System.in);
        int n = sc.nextInt();
        int arr[] = new int[n];
        for(int i = 0; i < n; i++){
            arr[i] = sc.nextInt();
        }

        ArrayList<Integer> ans = new ArrayList<>();

        int max = arr[n-1];
        ans.add(max);

        for(int i = n-2; i >= 0; i--){

            if(arr[i] > max){
                ans.add(arr[i]);
                max = arr[i];
            }
        }
        Collections.reverse(ans);
        System.out.println(ans);
```
