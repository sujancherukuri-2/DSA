<h2> 3.SPIRAL MATRIX TRAVERSAL</h2>

<h3><u><b>Approach</u></b></h3>

<p>RIGHT → DOWN → LEFT → UP<br>
top++ , right-- , bottom-- , left++</p>

<h4> TC - O(N * M) , SC - O(1)</h4>

```python
int[][] arr = {
            {1, 2, 3},
            {4, 5, 6},
            {7, 8, 9}
        };

        int r = arr.length;
        int c = arr[0].length;

        int top = 0;
        int bottom = r - 1;
        int left = 0;
        int right = c - 1;

        while(top <= bottom && left <= right){

            // left -> right
            for(int i = left; i <= right; i++){
                System.out.print(arr[top][i] + " ");
            }
            top++;

            // top -> bottom
            for(int i = top; i <= bottom; i++){
                System.out.print(arr[i][right] + " ");
            }
            right--;

            // right -> left
            if(top <= bottom){
                for(int i = right; i >= left; i--){
                    System.out.print(arr[bottom][i] + " ");
                }
                bottom--;
            }

            // bottom -> top
            if(left <= right){
                for(int i = bottom; i >= top; i--){
                    System.out.print(arr[i][left] + " ");
                }
                left++;
            }
        }
```
