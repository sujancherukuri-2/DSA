<h2> STOCK BUY AND SELL </h2>

<p>Input: prices = [7,1,5,3,6,4]<br>
Output: 5

Explanation: Buy on day 2 (price = 1) and sell on day 5 (price = 6), profit = 6-1 = 5.

Note: That buying on day 2 and selling on day 1 is not allowed because you must buy before you sell.</p>


<h3><u><b>1.Bruteforce</u></b></h3>

<p>
maxProfit = 0

For i = 0 → n-1

For j = i+1 → n-1

Compute profit = prices[j] - prices[i]

Update maxProfit</p>

<h4> TC - O(N^2) , SC - O(1)</h4>

```python
        Scanner sc = new Scanner(System.in);
        int n = sc.nextInt();
        int prices[] = new int[n];

        for(int i = 0; i < n; i++){
            prices[i] = sc.nextInt();
        }

        int maxProfit = 0;

        for(int i = 0; i < n; i++){
            for(int j = i + 1; j < n; j++){

                int profit = prices[j] - prices[i];

                if(profit > maxProfit){
                    maxProfit = profit;
                }
            }
        }

        System.out.println(maxProfit);
```

<h3><u><b>2. Optimal </u></b></h3>

<p>minPrice = prices[0]<br>

maxProfit = 0

Traverse array

profit = prices[i] - minPrice

Update maxProfit

minPrice = min(minPrice, prices[i])

Print maxProfit</p>

<h4> TC - O(N) , SC - O(1)</h4>

```python
        Scanner sc = new Scanner(System.in);
        int n = sc.nextInt();
        int prices[] = new int[n];

        for(int i = 0; i < n; i++){
            prices[i] = sc.nextInt();
        }

        int minPrice = prices[0];
        int maxProfit = 0;

        for(int i = 1; i < n; i++){

            int profit = prices[i] - minPrice;

            if(profit > maxProfit){
                maxProfit = profit;
            }

            if(prices[i] < minPrice){
                minPrice = prices[i];
            }
        }

        System.out.println(maxProfit);
```
