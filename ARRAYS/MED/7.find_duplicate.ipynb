<h2> 7.FIND DUPLICATE</h2>


<h3><u><b>1.Bruteforce</u></b></h3>

<p>Traverse array<br>

For every element arr[i]

Count its occurrences

If count > 1 → duplicate found</p>

<h4> TC - O(N^2) , SC - O(1)</h4>

```python
        Scanner sc = new Scanner(System.in);

        int n = sc.nextInt();   
        int arr[] = new int[n];

        for(int i = 0; i < n; i++){
            arr[i] = sc.nextInt();
        }

        for(int i = 0; i < n; i++){
            int count = 0;
            for(int j = 0; j < n; j++){
                if(arr[i] == arr[j]){
                    count++;
                }
            }
            if(count > 1){
                System.out.println(arr[i]);
                break;
            }
        }
```

<h3><u><b> 2.HASHING </u></b></h3>

<p> Create HashSet<br>

Traverse array

If element already in set → duplicate

Otherwise insert element</p>

<h4> TC - O(N) , SC - O(N)</h4>

```python
        Scanner sc = new Scanner(System.in);

        int n = sc.nextInt();
        int arr[] = new int[n];

        for(int i = 0; i < n; i++){
            arr[i] = sc.nextInt();
        }

        HashSet<Integer> set = new HashSet<>();
        for(int num : arr){
            if(set.contains(num)){
                System.out.println(num);
                break;
            }
            set.add(num);
        }
```

<h2> OPTIMAL </h2>

<p>3️⃣ Optimal Approach (Floyd's Cycle Detection)<br>

This is the most famous interview solution.

Idea based on Linked List Cycle Detection.

Why it works

Since numbers are 1 → n, they can be treated as indexes forming a cycle.

Duplicate creates a cycle.

Algorithm
Phase 1: Find meeting point
slow = nums[slow]
fast = nums[nums[fast]]

Repeat until they meet.

Phase 2: Find cycle start

Set slow = nums[0]

Move both one step

Where they meet = duplicate</p>

```python
        Scanner sc = new Scanner(System.in);

        int n = sc.nextInt();
        int nums[] = new int[n];

        for(int i = 0; i < n; i++){
            nums[i] = sc.nextInt();
        }

        int slow = nums[0];
        int fast = nums[0];

        // phase 1
        do{
            slow = nums[slow];
            fast = nums[nums[fast]];
        }while(slow != fast);

        // phase 2
        slow = nums[0];

        while(slow != fast){
            slow = nums[slow];
            fast = nums[fast];
        }

        System.out.println(slow);
```
