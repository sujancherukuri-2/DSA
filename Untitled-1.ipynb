```python
import java.util.Scanner;

Scanner sc = new Scanner(System.in);

// Input format:
// first number: array size n
// next n numbers: array elements
int n = sc.nextInt();
int countOnes = 0;

for (int i = 0; i < n; i++) {
    int value = sc.nextInt();
    if (value == 1) {
        countOnes++;
    }
}

System.out.println(countOnes);
```
